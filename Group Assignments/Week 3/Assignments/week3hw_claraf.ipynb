```python
# now going to try and clip only for neighborhoods chinatown, lincoln heights, elysian park, and echo park
neighborhoods = geopandas.read_file('LA_Times_Neighborhood_Boundaries.zip')
```

```python
neighborhoods.plot(figsize=(12,10))
```

```python
neighborhoods.sample(5)
```

```python
relevant_neighborhoods = set(('Echo Park','Lincoln Heights','Chinatown','Elysian Park'))
clipped_neighborhoods = neighborhoods.loc[neighborhoods['name'].isin(relevant_neighborhoods)] 
clipped_neighborhoods.head()
```

```python
tracts_rentage_clipped = geopandas.clip(tracts_rentage, clipped_neighborhoods)
```

```python
tracts_rentage_clipped.plot(figsize=(12,10))
```

```python
# now clipping original map to 4 neighborhood area only
import folium
tracts_rentage_clipped.plot(figsize=(12,10),
                 column='Renter 65 to 74 Years',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
tracts_rentage_clipped.plot(figsize=(12,10),
                 column='Renter 75 to 84 Years',
                 legend=True, 
                 scheme='NaturalBreaks')
```

```python
# creating choropleth map with the data and clipped maps

lat = tracts_rentage_clipped.geometry.centroid.x.mean()
long = tracts_rentage_clipped.geometry.centroid.y.mean()
print(lat,long)
```

```python
m = folium.Map(location=[long,lat], zoom_start=13,tiles='CartoDB positron')
m
```

```python
folium.Choropleth(
    geo_data = tracts_rentage_clipped,
    data = tracts_rentage_clipped,
    columns=['Geo_FIPS', 'Renter 65 to 74 Years'],
    key_on="feature.properties.Geo_FIPS",
    fill_color="BuPu",
    line_weight=0.1, 
    fill_opacity=0.8,
    line_opacity=0.2,
    legend_name="Renters 65 to 74 Years").add_to(m)
m
```
